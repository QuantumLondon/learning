```python
from qiskit import QuantumCircuit
```

```python
qc = QuantumCircuit(3, 3)
```

```python
qc.measure([0,1,2], [0,1,2])
```

```python
qc.draw('mpl')
```

Only answer $A$ works, because `measure()` takes two arguments: qubits, and target bits. Both of them in form of lists or tuples. Answer $C$ would create three new classical bits, instead of using the existing ones.
