```python
from qiskit import QuantumCircuit
from qiskit.circuit.library import CXGate
```

```python
qc1 = QuantumCircuit(3)
qc2 = QuantumCircuit(3)
qc3 = QuantumCircuit(3)
qc4 = QuantumCircuit(3)
```

Answer $A$

```python
qc1.ccx(0, 1, 2)
qc1.draw('mpl')
```

Answer $B$

```python
qc2.mct([0, 1], 2)
qc2.draw('mpl')
```

Answer $C$

```python
ccx = CXGate().control()
qc3.append(ccx, [0, 1, 2])
qc3.draw('mpl')
```

Answer $D$

```python
qc4.cry(0, 1, 2)
qc4.draw('mpl')
```

Correct answer: $D$
