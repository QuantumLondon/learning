```python
from qiskit import QuantumCircuit
```

```python
qc1 = QuantumCircuit(3, 3)
qc2 = QuantumCircuit(3, 3)
qc3 = QuantumCircuit(3, 3)
qc4 = QuantumCircuit(3, 3)
qc5 = QuantumCircuit(3, 3)
```

Answer $A$

```python
qc1.barrier(qc1)
qc1.draw('mpl')
```

Answer $B$

```python
qc2.barrier([0, 1, 2])
qc2.draw('mpl')
```

Answer $C$

```python
qc3.barrier()
qc3.draw('mpl')
```

Answer $D$

```python
qc4.barrier(3)
qc4.draw('mpl')
```

Answer $E$

```python
qc5.barrier_all()
qc5.draw('mpl')
```
