```python
from qiskit import QuantumCircuit
from qiskit import Aer
from qiskit import execute
from qiskit.visualization import plot_histogram
import math
```

```python
qc = QuantumCircuit(1)
```

```python
qc.ry((3 * math.pi)/4, 0)
```

We add measurement to the circuit, run it on the simulator, and then plot the results.

```python
qc.measure_all()
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(qc, simulator, shots=1000).result()
```

```python
counts = result.get_counts(qc)
```

```python
plot_histogram(counts)
```

Result of simulator is not ideal, but it's closest to answer $C$, which is the correct answer.
