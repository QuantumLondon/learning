```python
from qiskit import QuantumCircuit
from qiskit import Aer
from qiskit import execute
from qiskit.visualization import plot_histogram
```

```python
bell = QuantumCircuit(2)
```

```python
bell.h(0)
```

```python
bell.x(1)
```

```python
bell.cx(0, 1)
```

```python
bell.draw('mpl')
```

I add measure, run the simulation and plot the result of `bell circuit`.

```python
bell.measure_all()
```

```python
simulator = Aer.get_backend('qasm_simulator')
```

```python
result = execute(bell, simulator, shots=1000).result()
```

```python
counts = result.get_counts(bell)
```

```python
plot_histogram(counts)
```

Answer $A$ creates a bell state. Answer $B$ has gates in wrong order, answer $C$ has wrong control gate, and answer $D$ has no CNOT gate.
