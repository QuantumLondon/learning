```python
from qiskit import QuantumCircuit, QuantumRegister
```

```python
c1 = QuantumCircuit(4, 4)
c1.draw('mpl')
```

This one creates quantum circuit with 4 quantum, and 4 classical bits.

```python
c2 = QuantumCircuit(4)
c2.draw('mpl')
```

This one creates quantum circuit with 4 quantum bits.

```python
c3 = QuantumCircuit(QuantumRegister(4, 'qr0'), QuantumRegister(4, 'cr1'))
c3.draw('mpl')
```

This one creates quantum circuit with 8 quantum bits. Four named `qr0` and four named `cr1`.

```python
c4 = QuantumCircuit([4, 4])
c4.draw('mpl')
```
