```python
from qiskit import QuantumRegister, QuantumCircuit
```

```python
inp_reg = QuantumRegister(2, name='inp')
```

```python
ancilla = QuantumRegister(1, name='anc')
```

```python
qc1 = QuantumCircuit(inp_reg, ancilla)
qc2 = QuantumCircuit(inp_reg, ancilla)
qc3 = QuantumCircuit(inp_reg, ancilla)
```

Code $A$

```python
qc1.h(inp_reg)
qc1.x(ancilla)
qc1.draw('mpl')
```

We can use gates on all bits in circuit.

Code $B$

```python
qc2.h(inp_reg[0:2])
qc2.x(ancilla[0])
qc2.draw('mpl')
```

We can use slicers to specify to which bits the gates should be applied.

Code $D$

```python
qc3.h(inp_reg[0])
qc3.h(inp_reg[1])
qc3.x(ancilla[0])
qc3.draw('mpl')
```

We can specify each bit, and gate separately.

Correct answers are $A$, $B$, $D$
